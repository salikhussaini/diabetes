```python
#Import mathplotLib
import matplotlib.pyplot as plt
#Pandas
import pandas as pd
#Numpy
import numpy as np

#Import Train_test_split
from sklearn.model_selection import GridSearchCV, train_test_split

#Load Data
data = pd.read_csv('diabetic_data_1.csv')
data.head()
```

```python
x = data
y1 = data['readmitted_No']
y2 = data['readmitted_<30']
y3 = data['readmitted_>30']
y4 = data['readmitted']
y1 = pd.DataFrame(y1)
y2 = pd.DataFrame(y2)
y3 = pd.DataFrame(y3)
y4 = pd.DataFrame(y4)
```

```python
from sklearn.preprocessing import StandardScaler

x = data.drop(['readmitted_No' , 'readmitted_<30', 'readmitted_>30','readmitted'], axis = 1)
x = data[["Caucasian","AfricanAmerican","Asian","Hispanic","Other",
          "Unknown","Female","Male","Age_0-10","Age_10-20","Age_20-30","Age_30-40","Age_40-50","Age_50-60","Age_60-70","Age_70-80","Age_80-90","Age_98","admission_type_id","discharge_disposition_id","admission_source_id","time_in_hospital","num_lab_procedures","num_procedures","num_medications","number_outpatient","number_emergency","number_inpatient","number_diagnoses"]]
autoscaler = StandardScaler()
x = autoscaler.fit_transform(x)
x = pd.DataFrame(x)
x.head()
```

```python
#Split dataset into 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(x, y1, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
